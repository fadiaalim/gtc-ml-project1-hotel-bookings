```python
# ============================================
# GTC ML Project 1 - Hotel Bookings Data Cleaning & Preprocessing
# ============================================

# Phase 1: Exploratory Data Analysis (EDA)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno

# Load the dataset
df = pd.read_csv("hotel_bookings.csv")

# Quick look at the data
print(df.shape)
print(df.info())
print(df.describe(include="all").T)

# Missing values visualization
msno.matrix(df)
plt.show()

# Outlier detection with boxplots
plt.figure(figsize=(10,4))
sns.boxplot(x=df['adr'])
plt.title("ADR Outliers")
plt.show()

plt.figure(figsize=(10,4))
sns.boxplot(x=df['lead_time'])
plt.title("Lead Time Outliers")
plt.show()

# ============================================
# Phase 2: Data Cleaning
# ============================================

# Handle missing values
df['company'].fillna(0, inplace=True)
df['agent'].fillna(0, inplace=True)
df['country'].fillna(df['country'].mode()[0], inplace=True)
df['children'].fillna(df['children'].median(), inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Handle outliers (cap adr > 1000 at 1000)
df['adr'] = np.where(df['adr'] > 1000, 1000, df['adr'])

# Fix data types for date columns
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# ============================================
# Phase 3: Feature Engineering & Preprocessing
# ============================================

# Create new features
df['total_guests'] = df['adults'] + df['children'] + df['babies']
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['is_family'] = np.where((df['children'] + df['babies']) > 0, 1, 0)

# Drop leakage columns
df.drop(['reservation_status', 'reservation_status_date'], axis=1, inplace=True)

# Encode categorical variables
categorical_cols = ['meal', 'market_segment', 'distribution_channel', 'reserved_room_type', 'assigned_room_type', 'deposit_type', 'customer_type']
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Frequency encoding for high-cardinality column "country"
country_freq = df['country'].value_counts().to_dict()
df['country'] = df['country'].map(country_freq)

# Final dataset split
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2, random_state=42)

print("Train shape:", train.shape)
print("Test shape:", test.shape)

# Save cleaned data (optional)
train.to_csv("hotel_bookings_train.csv", index=False)
test.to_csv("hotel_bookings_test.csv", index=False)
```
